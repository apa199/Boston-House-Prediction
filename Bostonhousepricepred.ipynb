```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("HousingData.csv")
data.rename(columns={'MEDV':'PRICE'}, inplace=True)
```

```python
data.isnull().sum()
```

```python
data=data.fillna(data.mean())
```

```python
data.isnull().sum()
```

```python
data.corr()
```

```python
import seaborn as sns
sns.heatmap(data= data.corr(), annot=True )
```

```python
sns.regplot(x=data["RM"],y=data["PRICE"],data= data)
```

```python
x=data.iloc[::,:-1]
y=data.iloc[:,-1]
from sklearn.model_selection import train_test_split
X_test, X_train,y_test,y_train=train_test_split(x,y, test_size=0.3, random_state=20)


    

```

```python
x.shape
```

```python
X_test.head()
X_train.head()
```

```python
 #before scaling
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig,(ax1) = plt.subplots(ncols=1, figsize=(10,8))
sns.kdeplot(X_train, ax=ax1)
sns.kdeplot(X_test, ax=ax1)
```

```python
##Standardziation

from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
x_train=scaler.fit_transform(X_train)
x_test=scaler.fit_transform(X_test)
```

```python
x_train.head()
```

```python
fig,(ax1) = plt.subplots(ncols=1, figsize=(10,8))
sns.kdeplot(x_train, ax=ax1)
sns.kdeplot(x_test, ax=ax1)
```

```python
# Linear Model
from sklearn.linear_model import LinearRegression
regression= LinearRegression()
regression.fit(x_train, y_train)
```

```python
x_test
```

```python
Intercept= regression.intercept_
coeff=regression.coef_
print(coeff)
```

```python
regpred=regression.predict(x_test)
```

```python
plt.scatter(y_test, regpred)
```

```python
residue= y_test-regpred
sns.kdeplot(residue)
```

```python
## Evaluation of the model:
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
import numpy as np
mse= mean_squared_error(y_test, regpred)
mae= mean_absolute_error(y_test, regpred)
rmse= np.sqrt(mse)

print(mse)
print(mae)
print(rmse)
```

```python
from sklearn.metrics import r2_score
rsq= r2_score(y_test, regpred)
rsq
```

```python
## R2 adjusted
r2ad=1-(((1-rsq)*(len(y_test-1))/(len(y_test)-X_test.shape[1]-1)))
r2ad
```
